# Notebook responsável por centralizar as funções reutilizáveis

```python
def read_table(table):
    """
    Lê uma tabela de um banco de dados PostgreSQL e retorna um DataFrame do PySpark.

    Esta função estabelece uma conexão JDBC com um banco PostgreSQL local
    e carrega os dados da tabela especificada diretamente em um DataFrame Spark.

    Parâmetros:
    ----------
    table : str
        Nome da tabela a ser lida no banco de dados.        

    Retorno:
    -------    
        DataFrame contendo os todos os dados da tabela carregada do PostgreSQL.    
    """
    url = 'jdbc:postgresql://127.0.0.1:5432/sicooperative'
    
    df = (
        spark
        .read
        .format("jdbc")
        .option("driver", "org.postgresql.Driver")
        .option("url", url)
        .option("dbtable", table)
        .option("user", "postgres")
        .option("password", "postgres")
        .load()
    )
    return df
```
